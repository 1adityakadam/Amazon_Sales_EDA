```python
amazon_data = pd.read_csv('amazon_sale_report.csv')
amazon_data.head()
```

```python
amazon_df = pd.DataFrame(amazon_data)
amazon_df
```

```python
amazon_data.memory_usage()
```

```python
amazon_df.memory_usage(deep=True)
```

What's going on with the deep=True option? 

When you don't specify deep=True, the memory usage method just tells you the amount of memory used by the numpy arrays in the pandas dataframe. 

When you pass deep=True, it tells you the total amount of memory by including the memory used by all the text objects.

```python
# columns to convert into category data type to reduce memory

column_convert = amazon_df.columns[3:12]
amazon_df[column_convert] = amazon_df[column_convert].astype('category')
```

```python
# Define the columns to convert
columns_to_convert = ['Order ID', 'Courier Status', 'currency', 'ship-city', 'ship-state', 'ship-country', 'promotion-ids', 'fulfilled-by', 'Unnamed: 22']

# Convert the specified columns to category dtype
amazon_df[columns_to_convert] = amazon_df[columns_to_convert].astype('category')
```

```python
# Convert 'Date' column to datetime
amazon_df['Date'] = pd.to_datetime(amazon_df['Date'])



# Convert 'ship-postal-code' to integer, handling potential non-numeric values
amazon_df['ship-postal-code'] = pd.to_numeric(amazon_df['ship-postal-code']).astype('Int64')
```
